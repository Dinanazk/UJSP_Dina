```python
#Latihan K-Means Clustering

# Import Library
import pandas as pd
```

```python
# Load Dataset
df = pd.read_csv("Mall_Customers.csv")
df.head(3)
```

```python
# Mengubah nama kolom
df = df.rename(columns={'Gender': 'gender','Age': 'age','Annual Income (k$)': 'annual_income','Spending Score (1-100)': 'spending_score'})

# Mengubah data kategori menjadi data numerik
df['gender'] = df['gender'].replace({'Female': 0, 'Male': 1})

df.head()
```

```python
# Delete Kolom Customer id dan gender
X = df.drop(['CustomerID', 'gender'], axis=1)
```

```python
# Import Algoritma
from sklearn.cluster import KMeans
```

```python
# Membuat List dari nilai K 1-11 (inertia_)
clusters = []
for i in range(1, 11):
    km = KMeans(n_clusters=i).fit(X)
    clusters.append(km.inertia_)
```

```python
# import library untuk visualisasi
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# Membuat plot inertia
fig, ax = plt.subplots(figsize=(8,4))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
# membuat objek KMeans
km5 = KMeans(n_clusters=5).fit(X)

# menambahkan kolom Label pada dataset
X['Labels'] = km5.labels_
```

```python
# Membuat plot KMeans dengan 5 klaster
plt.figure(figsize=(8,4))
sns.scatterplot(data=X, x='annual_income', y='spending_score', hue='Labels',
                palette=sns.color_palette('hls', 5))
plt.title('KMeans dengan 5 Cluster')
plt.show()
```

```python
# Membuat daftar emoji dari Unicode
list_emoji = [e for e in emoji.EMOJI_DATA.keys()]
```

```python
# Menghitung jumlah judul yang mengandung emoji
count = 0
for em in list_emoji:
    for title in data.title:
        if em in title:
            count += 1
print("How many titles use emoji?", count)
```

```python
# Fungsi untuk menerjemahkan emoji dalam teks menjadi deskripsi teks
def demojize(text):
    for em in list_emoji:
        if em in text:
            en_text = emoji.demojize(em)
            text = text.replace(em, " " + en_text + " ")
    return text
```

```python
# Menerapkan fungsi demojize ke kolom title
data["title_emoji"] = data.title.apply(demojize)
```

```python
# Menemukan indeks judul yang mengandung emoji
title_with_emoji_idx = [
    idx for idx in range(len(data.title))
    for em in list_emoji
    if em in data.loc[idx, "title"]
]
```

```python
#Konversi ke DataFrame
train_tfidf = pd.DataFrame(train_tfidf.toarray(), columns=vectorizer.get_feature_names_out())
dev_tfidf = pd.DataFrame(dev_tfidf.toarray(), columns=vectorizer.get_feature_names_out())
```

```python
with pd.option_context("display.max_columns", 100):
    display(train_tfidf.sample(5))
```

```python
#Model Training and Evaluation
dict_models = {
    "logistic_regression": LogisticRegression(),
    "naive_bayes": MultinomialNB(),
    "svm": LinearSVC(),
    "decision_tree": DecisionTreeClassifier(random_state=11),
    "random_forest": RandomForestClassifier(random_state=11)
}
```

```python
for model in dict_models.values():
    print(f"-- {model.__class__.__name__} --")
    model.fit(train_tfidf, y_train)  # Melatih model
    y_pred = model.predict(dev_tfidf)  # Prediksi data validasi
    print("Reports on dev set:", classification_report(y_dev, y_pred), sep="\n")
```

```python
#LinearSVC
svm_grid_search = GridSearchCV(
    dict_models["svm"],
    {"C": (10, 1, .1, .05, .01)},
)

svm_grid_search.fit(train_tfidf, y_train)
svm_pred_dev = svm_grid_search.predict(dev_tfidf)

print("Reports on train set:",
    classification_report(
        y_train,
        svm_grid_search.predict(train_tfidf)
    ), sep="\n"
)

print("Reports on dev set:", classification_report(y_dev, svm_pred_dev), sep="\n")
```

```python
svm_grid_search.best_params_
```

```python
#Dataset (Corpus)

corpus = [
    'The Industrial Revolution 4.0 is changing most of the business activities',
    'The Industrial Revolution 4.0 has five technologies that are the main points',
    'Industry 4.0 opens new challenges for companies'
]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorizer_X = vectorizer.fit_transform(corpus).todense()
vectorizer_X
```

```python
vectorizer.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import euclidean_distances

for i in range(len(vectorizer_X)) :
    for j in range(i, len(vectorizer_X)):
        if i == j:
            continue
        jarak = euclidean_distances(np.asarray(vectorizer_X[i]), np.asarray(vectorizer_X[j]))
        print(f'Jarak dokumen {i+1} dan {j+1}: {jarak}')
```

```python
#Inisialisasi CountVectorizer dengan Stop Words

from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english')
vectorizer_X = vectorizer.fit_transform(corpus).todense()
vectorizer_X
```
